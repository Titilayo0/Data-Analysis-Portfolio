```python
#import packages to manipulate data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# Import the data
df = pd.read_csv("insurance.csv")
df.head()
```

```python
#extract the necessary column to build algorithm
df1 = df[['age', 'charges']]
df1.head()
```

```python
df1.shape
```

```python
#use scatter plot to explore relationship between age and insurance costs(charges)
x = df1.iloc[:,:1].values.reshape(-1,1)
y = df1.iloc[:,-1].values

plt.scatter(x,y,color = 'm')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
# Fit the model
insurance_model = LinearRegression()
insurance_model.fit(x,y)
```

```python
# Plot the data and fit the model
y_pred = insurance_model.predict(x)

plt.scatter(x,y,color = 'm')
plt.plot(x,y_pred,color = 'g')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
# Predict an unknown value
unknown_x = [[36]] 

x_pred = np.append(x, unknown_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

insurance_cost = insurance_model.predict(unknown_x)
rounded_insurance_cost = np.round(insurance_cost, 2)
print("The insurance cost for age 36:", rounded_insurance_cost)
```
